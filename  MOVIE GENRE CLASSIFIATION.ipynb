# Count Baseline

To start off with train a simple count baseline, to see how well we perform if we only look at the words of the training dataset.

```python
!pip3 install nltk
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
#printing out the stopwords in english
stopwords_en = stopwords.words("english")
stopwords_en
```

```python
# count the frequency of each stop word
stopword_freq = {}
for word in stopwords_en:
    stopword_freq[word] = stopword_freq.get(word, 0) + 1

# sort the stop words by frequency in descending order
stopword_freq_sorted = sorted(stopword_freq.items(), key=lambda x: x[1], reverse=True)

# extract the top 20 most common stop words and their frequencies
top_stopwords = stopword_freq_sorted[:20]
top_stopwords_words = [pair[0] for pair in top_stopwords]
top_stopwords_freq = [pair[1] for pair in top_stopwords]

# create a bar chart of the top 20 most common stop words
fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(top_stopwords_words, top_stopwords_freq)
ax.invert_yaxis()
ax.set_xlabel("Frequency")
ax.set_ylabel("Stop word")
ax.set_title("Top 20 most common stop words in English")
plt.show()
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
word_tokenize("This is just a test.")
```

```python
most_common_words = {}

for l in labels:
    tokenized_text = df_train[df_train["Genre"] == l]["Plot"].str.lower().apply(
        lambda x: [e for e in word_tokenize(x) if e not in stopwords_en and len(e) > 2]
    )
    
    most_common_words[l] = Counter(tokenized_text.explode()).most_common()
```

```python
#checking for the most common word in each genre and printing out the first 15 words
for k,v in most_common_words.items():
    print(k,"\n",v[:15])
    print("="*120)
```

```python
#checking the top 10,000 most common words
most_common_words_10k = {}
for k,v in most_common_words.items():
    most_common_words_10k[k] = [word for word,count in v][:10000]
#printing the first three words of the romance genre    
most_common_words_10k["romance"][:3], len(most_common_words_10k["romance"])
```

```python
idx = 3
df_val.iloc[idx]
```

```python
#looping through the genre in most common words and printing the genre name and total number of common words
for Genre, words in most_common_words_10k.items():
    print(Genre)
    print(len(list(set(word_tokenize(df_val.iloc[idx].Plot)) & set(words))))
    print("="*10)
```

```python
from random import randint
```

```python
def count_classifier(row):
    len_intersection = []
    
    for l in labels:
        len_intersection.append(len(list(set(word_tokenize(row)) & set(most_common_words_10k[l]))))

    winner = list(np.argwhere(len_intersection == np.amax(len_intersection)).flatten())
    
    if len(winner) == 1:
        pred = winner[0]
    else:
        pred = winner[randint(0,len(winner)-1)]
    
    return pred
```

```python
df_val["count_clf_pred"] = df_val["Plot"].apply(count_classifier)
df_val
```

```python
# From here: https://github.com/nlp-with-transformers/notebooks/blob/main/02_classification.ipynb

from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_preds, y_true, labels):
    cm = confusion_matrix(y_true, y_preds, normalize="true")
    fig, ax = plt.subplots(figsize=(6, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap="Blues", values_format=".2f", ax=ax, colorbar=False)
    plt.title("Normalized confusion matrix")
    plt.show()
```

```python
print(classification_report(list(df_val["label"]),df_val["count_clf_pred"], target_names=labels))
```

```python
plot_confusion_matrix(df_val["count_clf_pred"], list(df_val["label"]), labels)
```

# Transformer Model

Now we are training a more complex model, namely one based on the Transformer archicture. Some the code is based on the excellent NLP with Transformers book, in particular chapter 2: https://github.com/nlp-with-transformers/notebooks/blob/main/02_classification.ipynb.

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer

model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
def tokenize(batch):
    return tokenizer(batch["Plot"], padding=True, truncation=True)
```

```python
dataset_encoded = dataset.map(tokenize, batched=True, batch_size=None)
```

```python
print(dataset_encoded["train"].column_names)
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
from transformers import AutoModelForSequenceClassification, AutoConfig

num_labels = len(labels)

#using the Hugging Face Transformers library to load a pre-trained model for sequence classification with a specified number of labels.
config = (AutoConfig
          .from_pretrained(model_ckpt, num_labels=num_labels, 
                           label2id=label2id, id2label=id2label))

model = (AutoModelForSequenceClassification
         .from_pretrained(model_ckpt, config=config)
         .to(device))
```

```python
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    f1 = f1_score(labels, preds, average="weighted")
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "f1": f1}
```

```python
from transformers import Trainer, TrainingArguments

#setting up a training argument using a pre trained transformer model
batch_size = 16
logging_steps = len(dataset_encoded["train"]) // batch_size
model_name = f"{model_ckpt}-finetuned"
training_args = TrainingArguments(output_dir=model_name,
                                  num_train_epochs=5,
                                  learning_rate=2e-5,
                                  per_device_train_batch_size=batch_size,
                                  per_device_eval_batch_size=batch_size,
                                  weight_decay=0.01,
                                  evaluation_strategy="epoch",
                                  disable_tqdm=False,
                                  logging_steps=logging_steps,
#                                   push_to_hub=True, 
                                  log_level="error",
                                  save_total_limit=1)
```

```python
from transformers import Trainer

trainer = Trainer(model=model, args=training_args, 
                  compute_metrics=compute_metrics,
                  train_dataset=dataset_encoded["train"],
                  eval_dataset=dataset_encoded["valid"],
                  tokenizer=tokenizer)
trainer.train();
```

```python
preds_output = trainer.predict(dataset_encoded["valid"])
```

```python
preds_output.metrics
```

```python
y_preds = np.argmax(preds_output.predictions, axis=1)
y_valid = np.array(df_val["label"])
```

```python
plot_confusion_matrix(y_preds, y_valid, labels)
```

```python
print(classification_report(y_valid,y_preds, target_names=labels))
```

# Inference

```python
Plot = "Thor enlists the help of Valkyrie, Korg and ex-girlfriend Jane Foster to fight Gorr the God Butcher, who intends to make the gods extinct."

encoding = tokenizer(Plot,  padding=True, truncation=True, return_tensors="pt")
encoding = {k: v.to(trainer.model.device) for k,v in encoding.items()}

outputs = trainer.model(**encoding)
```

```python
logits = outputs.logits
logits.shape
```

```python
predictions = torch.nn.functional.softmax(logits.squeeze().cpu(), dim=-1)
predictions
```
